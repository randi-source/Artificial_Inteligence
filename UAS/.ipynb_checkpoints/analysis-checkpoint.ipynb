```python
import pandas as pd
import regex as re
from langdetect import detect

import nltk
from nltk.corpus import sentiwordnet as swn
from nltk.corpus import wordnet as wn
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('sentiwordnet')
```

```python
df = pd.read_csv("linkedinjobs.csv")
df.head()
```

```python
df.isna().sum()
```

```python
df_0 = df.dropna()
df_0.count()
```

```python
df_0.loc[:, "job_description"] = df_0["job_description"].replace("\s+", " ", regex=True)
```

```python
df_0.head()
```

```python
df_0.loc[:, "job_description"] = df_0["job_description"].replace("[^-9A-Za-z ]", "", regex=True)
df_0.head()
```

```python
df_0["job_description"]
```

```python
def detect_language(text):
    try:
        return detect(text)
    except:
        return None
```

```python
df_0['language'] = df_0['job_description'].apply(detect_language)
```

```python
df_0 = df_0[df_0['language'] !='id']
```

```python
df_0['job_description'] = df_0['job_description'].str.lower()
df_0['job_description']
```

```python
def tokenize_text(text):
    return word_tokenize(text)
```

```python
df_tokenized = df_0['job_description'].apply(tokenize_text)
df_tokenized
```

```python
stopwords = nltk.corpus.stopwords.words('english')
def remove_stopwords(tokens):
    return [token for token in tokens if token not in stopwords]
```

```python
df_stopwords = df_tokenized.apply(remove_stopwords)
df_stopwords
```

```python
lemmatizer = nltk.WordNetLemmatizer()
def lemmatize_tokens(tokens):
    return [lemmatizer.lemmatize(token) for token in tokens]
```

```python
df_lemma = df_stopwords.apply(lemmatize_tokens)
df_lemma
```

```python
from collections import Counter

# Flatten the list of tokens
tokens = [token for tokens in df_lemma for token in tokens]

# Count the frequency of each token
token_counts = Counter(tokens)

# Print the most common tokens
print(token_counts.most_common())
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a WordCloud object
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = None, 
                min_font_size = 10).generate_from_frequencies(token_counts)

# Plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show()
```

```python
import plotly.express as px
import pandas as pd

# Convert the token counts to a DataFrame
df_plot = pd.DataFrame.from_dict(token_counts, orient='index', columns=['count'])

# Create the word cloud
fig = px.treemap(df_plot.reset_index(), path=['index'], values='count')

# Show the plot
fig.show()
```
