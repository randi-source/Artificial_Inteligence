## Topic Recognition

```python
import gensim
from gensim import corpora
```

```python
df_1["job_description_bigram_2"] = [
    [" ".join(pair) for pair in bigram_list]
    for bigram_list in df_1["job_description_bigram"]
]
```

```python
dictionary = corpora.Dictionary(df_1["job_description_bigram_2"].values)
```

```python
doc_term_matrix = [dictionary.doc2bow(doc) for doc in df_1["job_description_bigram_2"].values]
```

```python
Lda = gensim.models.ldamodel.LdaModel
```

```python
ldamodel = Lda(doc_term_matrix, num_topics=3, id2word = dictionary, passes=50)
```

```python
# Get the number of words in your vocabulary
num_words = len(ldamodel.id2word)

# Get all the words for each topic
topics = ldamodel.show_topics(num_topics=3, num_words=num_words, formatted=False)

# Create a dictionary with each topic and its words
topic_dict = {}
for topic_id, topic in topics:
    topic_dict[topic_id] = [word for word, _ in topic]
```

```python
topics
```

```python
from wordcloud import WordCloud

# For each topic, create a word cloud
for i, topic in topics:
    # Create a dictionary of word frequencies
    word_freq = dict(topic)

    # Create a word cloud
    wordcloud = WordCloud(width=1000, height=500, 
                          collocations=False,  # This line takes care of not repeating bi-grams
                          ).generate_from_frequencies(word_freq)

    # Display the word cloud
    plt.figure(figsize=(15,8))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.title(f'Topic {i}')
    plt.show()
```
